## 분산 분석(비모수 검정) - Mann-Whitney U Test 사용

### 두 그룹의 평균간에 유의미한 차이가 있는가??

이때,
대립가설(H1) : 두 그룹간 유의한 차이가 있다.

귀무가설(H0) : 두 그룹간 유의한 차이가 없다.

### p-value 가 0.05보다 작으므로 귀무가설 기각. 대립가설 채택 즉, 두 그룹간 유의한 차이가 있다.

### 두 그룹의 평균간에 유의미한 차이가 있는가??

### p-value 가 0.05보다 작으므로 귀무가설 기각. 대립가설 채택. 즉, 두 집단간의 유의한 차이가 있다.

## Q) 노선별로 역별/요일별 승하차승객수를 비교해 볼 수 있을까? (1~9호선, 역별/요일별 heatmap)

```python
raw['노선명'].unique()

#노선명 정보 확인
```

```python
line_list = []
for line in raw['노선명'].unique():
    if line[1:] == '호선':
        line_list.append(line)
line_list

# N호선만 추출 (1 ~ 9 한자리수)
```

```python
for line in sorted(line_list):
    
    data_line = raw[raw['노선명'] == line]
    df_pivot = data_line.pivot_table(index = ['역명'], columns = '요일', values = '승하자총승객수',aggfunc = 'sum')
    df_pivot = df_pivot[['월','화','수','목','금','토','일']]
    df_pivot = df_pivot / 10000  
    #만명단위로 시각화
    
    fig, ax = plt.subplots( figsize=(6,len(df_pivot)/3 ) )   # 그래프 사이즈를 조정하여, 역 수가 많은 경우는 세로를 길게 표현
    plt.title(f"{line} 역별/요일별 승객수", fontsize = 20)
    sns.heatmap(df_pivot, cmap = "Blues", annot = True, fmt = '.1f')
```
