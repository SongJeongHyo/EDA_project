# 칼럼 추가를 먼저 (코로나, 주말 칼럼)

```python
df17_cop = df17.copy()
df18_cop = df18.copy()
df19_cop = df19.copy()
df20_cop = df20.copy()
df21_cop = df21.copy()
```

```python
df17_cop
```

```python
df17_cop['기준일ID'] = pd.to_datetime(df17_cop['기준일ID'], format='%Y%m%d')
df18_cop['기준일ID'] = pd.to_datetime(df18_cop['기준일ID'], format='%Y%m%d')
df19_cop['기준일ID'] = pd.to_datetime(df19_cop['기준일ID'], format='%Y%m%d')
df20_cop['기준일ID'] = pd.to_datetime(df20_cop['기준일ID'], format='%Y%m%d')
df21_cop['기준일ID'] = pd.to_datetime(df21_cop['기준일ID'], format='%Y%m%d')
```

```python
df_total = pd.concat([df17_cop,df18_cop,df19_cop,df20_cop,df21_cop])
```

```python
df_total['year'] = df_total['기준일ID'].dt.year 
df_total['month'] = df_total['기준일ID'].dt.month 
df_total['day'] = df_total['기준일ID'].dt.day 
df_total['dayofweek'] = df_total['기준일ID'].dt.dayofweek #요일 (월 = 0)
```

```python
df_total
```

```python
df_total = df_total.reset_index()
```

```python
df_total['Holiday'] = 0
for i in range(0,len(df_total)):
    if df_total['dayofweek'][i] > 4:
        df_total['Holiday'][i] = 1
```

```python
df_total = df_total.drop('index',axis=1)
```

```python
df_total['Holid'] = 0
```

```python
df_total
```

```python
df_total.loc[(df_total['year']==2017) & (df_total['month']==1) & (df_total['day'].between(27, 30)),'Holid'] = 1
df_total.loc[(df_total['year']==2017) & (df_total['month']==10) & (df_total['day'].between(3, 9)),'Holid'] = 1

df_total.loc[(df_total['year']==2018) & (df_total['month']==2) & (df_total['day'].between(15, 18)),'Holid'] = 1
df_total.loc[(df_total['year']==2018) & (df_total['month']==9) & (df_total['day'].between(22, 26)),'Holid'] = 1

df_total.loc[(df_total['year']==2019) & (df_total['month']==2) & (df_total['day'].between(2, 6)),'Holid'] = 1
df_total.loc[(df_total['year']==2019) & (df_total['month']==9) & (df_total['day'].between(12, 15)),'Holid'] = 1

df_total.loc[(df_total['year']==2020) & (df_total['month']==1) & (df_total['day'].between(24, 27)),'Holid'] = 1
df_total.loc[(df_total['year']==2020) & (df_total['month']==9) & (df_total['day'].between(30, 30)),'Holid'] = 1
df_total.loc[(df_total['year']==2020) & (df_total['month']==10) & (df_total['day'].between(1, 4)),'Holid'] = 1

df_total.loc[(df_total['year']==2021) & (df_total['month']==2) & (df_total['day'].between(11, 14)),'Holid'] = 1
df_total.loc[(df_total['year']==2021) & (df_total['month']==9) & (df_total['day'].between(18, 22)),'Holid'] = 1
```
